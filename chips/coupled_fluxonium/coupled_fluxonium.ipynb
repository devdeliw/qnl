```python
from qiskit_metal import designs, MetalGUI, draw, Dict
from qiskit_metal.toolbox_metal.parsing import parse_value
```

```python
from components.right_flux_singlet import right_flux_singlet
from components.right_flux_doublet import right_flux_doublet
from components.coupling_resonator_ import resonator 
from components.readout_bus import readout_bus 
import numpy as np 
```

```python
design = designs.DesignPlanar() 
design.overwrite_enabled = True 

%load_ext autoreload 
%autoreload 2

gui = MetalGUI(design) 
```

```python
design.delete_all_components()
idc_pos_x = parse_value('0um', Dict())
idc_pos_y = parse_value('-3430um', Dict()) 
readout_bus(design, pos_x=idc_pos_x, pos_y=idc_pos_y) 
gui.rebuild() 
```

```python
right_flux_x = idc_pos_x + parse_value('1700um', Dict()) 
right_flux_y = idc_pos_y + parse_value('4242um', Dict()) 
nodes = right_flux_doublet(
    design, 
    pos_x=right_flux_x, 
    pos_y=right_flux_y,
) 
gui.rebuild() 
```

```python
pos_upper_x, pos_upper_y = nodes['upper_qubit_claw'] 
pos_lower_x, pos_lower_y = nodes['lower_qubit_claw'] 

resonator(design, pos_upper_x, pos_upper_y, key='upper_left', name='upper_doublet_resonator')
resonator(design, pos_lower_x, pos_lower_y, key='lower_left', name='lower_doublet_resonator')
gui.rebuild() 
```

```python
right_flux_sing_x = right_flux_x 
right_flux_sing_y = right_flux_y + parse_value('-2575um', Dict())

nodes = right_flux_singlet(
    design, 
    pos_x=right_flux_sing_x, 
    pos_y=right_flux_sing_y, 
    name='flux_singlet'
)
gui.rebuild() 
```

```python
sing_res_x, sing_res_y = nodes['qubit_claw'] 
resonator(design, sing_res_x, sing_res_y, key='upper_left', name='singlet_resonator')
```

```python
gui.rebuild() 
gui.autoscale()
```
