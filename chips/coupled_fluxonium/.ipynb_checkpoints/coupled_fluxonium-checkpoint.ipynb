```python
from qiskit_metal import designs, MetalGUI, draw, Dict
from qiskit_metal.toolbox_metal.parsing import parse_value
from qiskit_metal.qlibrary.terminations.launchpad_wb import LaunchpadWirebond
from qiskit_metal.qlibrary.terminations.short_to_ground import ShortToGround
from qiskit_metal.qlibrary.terminations.open_to_ground import OpenToGround
from qiskit_metal.qlibrary.tlines.pathfinder import RoutePathfinder
from qiskit_metal.qlibrary.QNLMetal.alignmentmarker import AlignmentMarker
```

```python
# in order of being made 
from components.right_flux_singlet import right_flux_singlet
from components.right_flux_doublet import right_flux_doublet
from components.coupling_resonator_ import resonator 
from components.readout_bus import readout_bus 
from components.left_transmon import left_transmon_resonator
from components.left_flux_doublet import left_flux_doublet
from collections import OrderedDict
import numpy as np 
```

```python
design = designs.DesignPlanar() 
design.overwrite_enabled = True 

%load_ext autoreload 
%autoreload 2

gui = MetalGUI(design) 
```

#### Right Fluxonium Doublet 

```python
design.delete_all_components()

right_flux_x = 0 
right_flux_y = 0
right_doublet_nodes = right_flux_doublet(
    design, 
    pos_x=right_flux_x, 
    pos_y=right_flux_y,
) 
gui.rebuild() 
```

#### Doublet Coupling Resonators

```python
pos_upper_x, pos_upper_y = right_doublet_nodes['upper_qubit_claw'] 
pos_lower_x, pos_lower_y = right_doublet_nodes['lower_qubit_claw'] 

nodes_doublet_upper_res = resonator(design, pos_upper_x, pos_upper_y, key='upper_doublet', name='upper_doublet_resonator')
nodes_doublet_lower_res = resonator(design, pos_lower_x, pos_lower_y, key='lower_doublet', name='lower_doublet_resonator')
gui.rebuild() 
```

#### Right Fluxonium Singlet

```python
right_flux_sing_x = right_flux_x 
right_flux_sing_y = right_flux_y + parse_value('-2575um', Dict())

right_singlet_nodes = right_flux_singlet(
    design, 
    pos_x=right_flux_sing_x, 
    pos_y=right_flux_sing_y, 
    name='flux_singlet'
)
gui.rebuild() 
```

#### Singlet Coupling Resonator

```python
sing_res_x, sing_res_y = right_singlet_nodes['qubit_claw'] 

singlet_resonator_nodes = resonator(design, sing_res_x, sing_res_y, key='singlet', name='singlet_resonator')
gui.rebuild() 
```

#### Readout Bus & InlineIDC

```python
idc_pos_x = nodes_doublet_upper_res.closest_to_bus[0] - parse_value('32.5um', Dict()) - parse_value('20.5um', Dict())
idc_pos_y = right_flux_y - parse_value('4242um', Dict())

readout_bus_nodes = readout_bus(design, pos_x=idc_pos_x, pos_y=idc_pos_y) 
gui.rebuild() 
```

#### Left Transmon & Resonator

```python
tsmn_x = right_flux_x - parse_value('3405um', Dict()) + parse_value('280um', Dict())
tsmn_y = right_flux_y + parse_value('1687um', Dict())

left_tsmn_nodes = left_transmon_resonator(design, tsmn_x, tsmn_y, name='left_transmon_resonator')
gui.rebuild() 
```

#### Left Fluxonium Doublet 

```python
left_flux_y = right_flux_y - parse_value('649.5um', Dict()) 
left_flux_x = tsmn_x

left_doublet_nodes = left_flux_doublet(
    design, 
    pos_x=left_flux_x, 
    pos_y=left_flux_y,
) 
gui.rebuild()
```

#### Left Doublet Resonators

```python
pos_upper_x, pos_upper_y = left_doublet_nodes['upper_qubit_claw'] 
pos_lower_x, pos_lower_y = left_doublet_nodes['lower_qubit_claw'] 

nodes_left_doublet_upper_res = resonator(design, pos_upper_x, pos_upper_y, key='upper_left_doublet', name='upper_left_doublet_resonator')
nodes_left_doublet_lower_res = resonator(design, pos_lower_x, pos_lower_y, key='lower_left_doublet', name='lower_left_doublet_resonator')
gui.rebuild()
```

#### Corner Launchpads that connect to Fluxlines

```python
def place_launchpad_cpw_fluxline(key):
    fillet = '200um'
    launch_pad_ops = {
        'orientation': '90',
        'lead_length': '0um',
        'pad_width': '375um',
        'pad_height': '125um',
        'taper_height': '100um',
        'trace_gap': '5um',
        'trace_width': '8um',
    }
    
    top_right = (
        right_flux_x + parse_value('2300um', Dict()),
        right_flux_y + parse_value('3706.5um', Dict())
    )
    
    if key == 'top_right':
        launch_pad_ops.update({
            'orientation': '270',
            'pos_x': top_right[0],
            'pos_y': top_right[1]
        })
    elif key == 'top_left':
        launch_pad_ops.update({
            'orientation': '270',
            'pos_x': top_right[0] - parse_value('8000um', Dict()),
            'pos_y': top_right[1]
        })
    elif key == 'bottom_right':
        launch_pad_ops.update({
            'orientation': '180',
            'pos_x': top_right[0] + parse_value('518.5um', Dict()),
            'pos_y': top_right[1] - parse_value('8518um', Dict())
        })
    elif key == 'bottom_left':
        launch_pad_ops.update({
            'orientation': '90',
            'pos_x': top_right[0] - parse_value('8000um', Dict()),
            'pos_y': top_right[1] - parse_value('9037um', Dict())
        })
    elif key == 'left_center':
        launch_pad_ops.update({
            'orientation': '0',
            'pos_x': top_right[0] - parse_value('8518.5um', Dict()),
            'pos_y': top_right[1] - parse_value('4518.5um', Dict())
        })
    else:
        raise ValueError(f"Unknown key: {key}")
    
    launch_pad = LaunchpadWirebond(design, name=f'{key}_launch_fluxline', options=launch_pad_ops)
    
    anchors = OrderedDict()
    if key == 'top_right':
        end_pos = right_doublet_nodes.upper_flux_line_end
        end_orientation = '270'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] - parse_value(fillet, Dict()) - parse_value('600um', Dict()),
            launch_pad_ops['pos_y'] - parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([-parse_value(fillet, Dict()), -parse_value('2200um', Dict())])
        anchors[2] = anchors[1] + np.array([-parse_value(fillet, Dict()) - parse_value('900um', Dict()),
                                            -parse_value(fillet, Dict())])
    elif key == 'top_left':
        end_pos = left_doublet_nodes.upper_flux_line_end
        end_orientation = '270'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] + parse_value(fillet, Dict()) + parse_value('400um', Dict()),
            launch_pad_ops['pos_y'] - parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([parse_value(fillet, Dict()), -parse_value('2500um', Dict())])
        anchors[2] = anchors[1] + np.array([parse_value(fillet, Dict()) + parse_value('1100um', Dict()),
                                            -parse_value(fillet, Dict())])
    elif key == 'bottom_right':
        end_pos = right_doublet_nodes.lower_flux_line_end
        end_orientation = '90'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] - parse_value(fillet, Dict()) - parse_value('500um', Dict()),
            launch_pad_ops['pos_y'] + parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([-parse_value(fillet, Dict()), parse_value('3000um', Dict())])
        anchors[2] = anchors[1] + np.array([-parse_value(fillet, Dict()) - parse_value('1718.5um', Dict()),
                                            parse_value(fillet, Dict())])
    elif key == 'bottom_left':
        end_pos = left_doublet_nodes.lower_flux_line_end
        end_orientation = '90'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] + parse_value(fillet, Dict()) + parse_value('500um', Dict()),
            launch_pad_ops['pos_y'] + parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([parse_value(fillet, Dict()), parse_value('3256.5um', Dict())])
        anchors[2] = anchors[1] + np.array([parse_value(fillet, Dict()) + parse_value('1000um', Dict()),
                                            parse_value(fillet, Dict())])
    elif key == 'left_center':
        fillet = '81um'
        end_pos = left_doublet_nodes.central_flux_line_end
        end_orientation = '0'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] + parse_value('100um', Dict()),
            launch_pad_ops['pos_y']
        ])
    else:
        raise ValueError(f"Invalid key: {key}")
    
    end_stg = ShortToGround(design, f'{key}_fl_cpw_end', options=Dict(
        pos_x=end_pos[0], pos_y=end_pos[1], orientation=end_orientation))
    
    cpw_ops = Dict(pin_inputs=Dict(
        start_pin=Dict(component=launch_pad.name, pin='tie'),
        end_pin=Dict(component=end_stg.name, pin='short')
    ))
    cpw_ops.trace_width = '8um'
    cpw_ops.trace_gap   = '5um'
    cpw_ops.fillet      = fillet
    cpw_ops.anchors     = anchors
    
    route = RoutePathfinder(design, f'{key}_fl_cpw', cpw_ops)
```

```python
place_launchpad_cpw_fluxline('top_right')  
place_launchpad_cpw_fluxline('top_left') 
place_launchpad_cpw_fluxline('bottom_right')  
place_launchpad_cpw_fluxline('bottom_left') 
place_launchpad_cpw_fluxline('left_center')
gui.rebuild() 
```

#### Corner Launchpads that connect to qubits

```python
def place_launchpad_cpw_qubit(key): 
    fillet = '180um' 
    launch_pad_ops = {
        'orientation': '0',
        'lead_length': '0um',
        'pad_width': '375um',
        'pad_height': '125um',
        'taper_height': '100um',
        'trace_gap': '5um',
        'trace_width': '8um',
    }
    
    top_right = (
        right_flux_x + parse_value('2818.5um', Dict()),
        right_flux_y + parse_value('3188um', Dict())
    )

    if key == 'top_right':
        launch_pad_ops.update({
            'orientation': '180',
            'pos_x': top_right[0],
            'pos_y': top_right[1]
        })
    elif key == 'top_left':
        launch_pad_ops.update({
            'orientation': '0',
            'pos_x': top_right[0] - parse_value('9037um', Dict()),
            'pos_y': top_right[1]
        })
    elif key == 'bottom_right':
        launch_pad_ops.update({
            'orientation': '90',
            'pos_x': top_right[0] - parse_value('518.5um', Dict()),
            'pos_y': top_right[1] - parse_value('8518um', Dict())
        })
    elif key == 'bottom_left':
        launch_pad_ops.update({
            'orientation': '0',
            'pos_x': top_right[0] - parse_value('9037um', Dict()),
            'pos_y': top_right[1] - parse_value('8000um', Dict())
        })
    elif key == 'right_center':
        launch_pad_ops.update({
            'orientation': '180',
            'pos_x': top_right[0],
            'pos_y': top_right[1] - parse_value('4000um', Dict())
        })
    else:
        raise ValueError(f"Unknown key: {key}")

    launch_pad_qubit = LaunchpadWirebond(design, name=f'{key}_launch_qubit', options=launch_pad_ops)

    anchors = OrderedDict()
    if key == 'top_right':
        end_pos_ = right_doublet_nodes.upper_right
        end_pos_ = end_pos_ + [parse_value('80um', Dict()), 0] 
        end_orientation = '180'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] - parse_value(fillet, Dict()) - parse_value('188.5um', Dict()),
            launch_pad_ops['pos_y'] - parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([0, -parse_value('2665um', Dict())])

    elif key == 'top_left':
        end_pos_ = left_doublet_nodes.upper_left 
        end_pos_ = end_pos_ - [parse_value('80um', Dict()), 0]
        end_orientation = '0'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] + parse_value(fillet, Dict()) + parse_value('188.5um', Dict()),
            launch_pad_ops['pos_y'] - parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([0, -parse_value('3215um', Dict())])
        
    elif key == 'bottom_right':
        end_pos_ = right_singlet_nodes.right 
        end_pos_ = end_pos_ + [parse_value('75um', Dict()), 0]
        end_orientation= '180'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] - 1.01*parse_value(fillet, Dict()),
            launch_pad_ops['pos_y'] + 1.01*parse_value(fillet, Dict())
        ])

        anchors[1] = anchors[0] - [parse_value('500um', Dict()), 0]
        anchors[2] = anchors[1] + np.array([0, parse_value('2053.5um', Dict())])
       
    elif key == 'bottom_left':
        end_pos_ = left_doublet_nodes.lower_left 
        end_pos_ = end_pos_ - [parse_value('80um', Dict()), 0]
        end_orientation = '0'
        anchors[0] = np.array([
            launch_pad_ops['pos_x'] + parse_value(fillet, Dict()) + parse_value('188.5um', Dict()),
            launch_pad_ops['pos_y'] + parse_value(fillet, Dict())
        ])
        anchors[1] = anchors[0] + np.array([0, parse_value('3540um', Dict())])
        
    elif key == 'right_center':
        end_pos_ = right_doublet_nodes.lower_right 
        end_pos_ = end_pos_ + [parse_value('80um', Dict()), 0]
        end_orientation = '180'
        anchors[0] = np.array([ 
            launch_pad_ops['pos_x'] - parse_value('188.5um', Dict()),
            launch_pad_ops['pos_y']
        ])
        anchors[1] = anchors[0] + np.array([0, parse_value('290um', Dict())])
    else:
        raise ValueError(f"Invalid key: {key}")
        
    end_otg_qubit = OpenToGround(design, f'{key}_qubit_cpw_end', options=Dict(
        pos_x=end_pos_[0], pos_y=end_pos_[1], orientation=end_orientation, 
        gap='0um', termination_gap='5um', width='18um'))

    cpw_ops_qubit = Dict(pin_inputs=Dict(
        start_pin=Dict(component=launch_pad_qubit.name, pin='tie'),
        end_pin=Dict(component=end_otg_qubit.name, pin='open')
    ))
    cpw_ops_qubit.trace_width = '8um'
    cpw_ops_qubit.trace_gap   = '5um'
    cpw_ops_qubit.fillet      = fillet
    cpw_ops_qubit.anchors     = anchors
    
    route = RoutePathfinder(design, f'{key}_qubit_cpw', cpw_ops_qubit)
```

```python
place_launchpad_cpw_qubit('top_right')  
place_launchpad_cpw_qubit('top_left')
place_launchpad_cpw_qubit('bottom_left')
place_launchpad_cpw_qubit('bottom_right')
place_launchpad_cpw_qubit('right_center')
gui.rebuild()
```

#### Alignment Markers 

```python
def place_markers(key): 

    pos_x = right_flux_x+parse_value('2500um', Dict())  
    pos_y = right_flux_y+parse_value('3388um', Dict())
    
    if key == 'upper_right': 
        pos_x, pos_y = pos_x, pos_y
    elif key == 'upper_left': 
        pos_x-=parse_value('8400um', Dict())
    elif key == 'lower_right': 
        pos_y-=parse_value('8400um', Dict()) 
    elif key == 'lower_left': 
        pos_x-=parse_value('8400um', Dict())
        pos_y-=parse_value('8400um', Dict()) 
    else: 
        raise ValueError(f"Invalid Key: {key}") 

    AlignmentMarker(design, name=f'{key}_marker', options=Dict(
        pos_x=pos_x, 
        pos_y=pos_y, 
        size='20um', 
    ))
```

```python
place_markers('upper_right') 
place_markers('lower_right') 
place_markers('upper_left') 
place_markers('lower_left')
```

```python
gui.rebuild() 
gui.autoscale()
```
